```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen
import html5lib
```

```python
!conda install lxml --yes
```

```python
link= "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
from IPython.display import display_html
display_html(link, raw=True)
```

```python
dfs = pd.read_html(link)
dfs
```

```python
df=dfs[0]
df
```

```python
empty = 'Not assigned'
df = df[(df.Postcode != empty ) & (df.Borough != empty) & (df.Neighbourhood != empty)]
```

```python
df.reset_index()
```

```python
def Neighbourhood_list(grouped):    
    return ', '.join(sorted(grouped['Neighbourhood'].tolist()))
                    
grp = df.groupby(['Postcode', 'Borough'])
df = grp.apply(Neighbourhood_list).reset_index(name='Neighbourhood')
```
